```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import seaborn as sns
import numpy as np

# dataset
iris = load_iris()
x, y = iris.data, iris.target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# models
model1 = LogisticRegression(max_iter=1000)
model2 = DecisionTreeClassifier()

# Train models
model1.fit(x_train, y_train)
model2.fit(x_train, y_train)

# predictions
y_pred1 = model1.predict(x_test)
y_pred2 = model2.predict(x_test)

# Evalution
acc1 = accuracy_score(y_test, y_pred1)
acc2 = accuracy_score(y_test, y_pred2)
print(f"Accuracy of Logistic Regression: {acc1:.4f}")
print(f"Accuracy of Decision Tree: {acc2:.4f}")

# Feature Importance for Logistic Regression (using coefficients)
logreg_importance = np.abs(model1.coef_[0])
# Feature Importance 
tree_importance = model2.feature_importances_

# Create a bar plot for feature importance
features = iris.feature_names
#log feature imp plot
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.barplot(x=logreg_importance, y=features,hue=logreg_importance,palette='pastel')
plt.title('Feature Importance - Logistic Regression',fontsize=16)
plt.xlabel('Importance',fontsize=22)

#decision feature imp plot 
plt.subplot(1,2,2)
sns.barplot( x=tree_importance, y=features,hue=tree_importance,palette='pastel')
plt.title('Feature Importance - Decision Tree',fontsize=16)
plt.xlabel('Importance', fontsize=22)

plt.tight_layout()
plt.show()
```
